```python
url = ''

headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:80.0) Gecko/20100101 Firefox/80.0',
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',}
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import random
import time
```

```python
res = requests.get(url, headers=headers)
res.raise_for_status()
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
list_length = len(soup.pre.find_all('a'))
list_of_files = soup.pre.find_all('a')
list_of_urls = []

for i in range(list_length):
    if (i > 4) and (i%2==1):
        list_of_urls.append(url + list_of_files[i]['href'])
        print(list_of_files[i]['href'])
```

```python
list_of_content = []
list_of_book_names = []
count =0

for url in list_of_urls:
    #Getting book content
    res2 = requests.get(url, headers=headers)
    list_of_content.append(res2)
#     time.sleep(random.random())
    #Getting book names
    parser = urlparse(url)
    list_of_book_names.append(parser.geturl().split('/')[-1].replace('%20',' '))
    print(list_of_book_names[count])
    with open(list_of_book_names[count], 'wb') as f:
        f.write(list_of_content[count].content)
    print('Saved')
    count += 1
```

```python
for book, content in list_of_book_names, list_of_content:
    with open(book, 'wb') as f:
        f.write(content.content)
```
